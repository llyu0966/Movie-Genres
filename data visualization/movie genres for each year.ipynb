```python
data = pd.read_csv('movies.csv')
```

```python
# show the general distribution of number of movies per year.
```

```python
y = data['Total']
x = data['Year']

plt.xlabel('Year')
plt.ylabel('Total')
```

```python
plt.scatter(x,y)
```

```python
plt.plot(data.Year, data.Total)

plt.show()
```

```python
plt.title('Number of Movies for per Genre(per Year)')

plt.plot(data.Year, data.Total, label='Total')
plt.plot(data.Year, data.Action, label='Action')
plt.plot(data.Year, data['Film_Noir'], label='Film_Noir')
plt.plot(data.Year, data.Adventure, label='Adventure')
plt.plot(data.Year, data.Horror, label='Horror')

plt.xticks(data.Year)

plt.legend()

plt.show()
```

```python
plt.title('Number of Movies for per Genre(per Year)')

plt.plot(data.Year, data.Total, label='Total')
plt.plot(data.Year, data.Action, label='Action')
plt.plot(data.Year, data['Film_Noir'], label='Film_Noir')
plt.plot(data.Year, data.Adventure, label='Adventure')
plt.plot(data.Year, data.Horror, label='Horror')

# iterator every decade
print(data.Year[::10])

plt.xticks(data.Year[::10])

plt.legend()

plt.show()
```

```python
plt.title('Number of Movies for per Genre(per Year)')

for genre in data:
    if genre != 'Year':
        if genre != 'Total':
          plt.plot(data.Year, data[genre], marker = '.', label = genre)
    
# iterator every decade
print(data.Year[::10])

plt.xticks(data.Year[::10])

plt.legend()

plt.show()
```

```python
bins = [0,50,100,150,200,250,300,350,400,450,500,550,600,650,700,750,800]

plt.hist(data.Total, bins=bins)

plt.xticks(bins)

plt.ylabel('Number of Years')
plt.xlabel('Total Number of Movies Each Year')

plt.show()
```
